```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
import warnings
warnings.filterwarnings('ignore')



driver=webdriver.Chrome()
driver.get("https://www.youtube.com/watch?v=eLS5GlhjpcA")
time.sleep(2)



driver = webdriver.Chrome()
driver.get("https://www.youtube.com/watch?v=eLS5GlhjpcA")
time.sleep(2)

# Lists to store comments, upvotes, and times
comments = []
upvotes = []
timestamps = []

# Scroll until at least 500 comments are loaded
while len(comments) <= 500 or len(upvotes) <= 500 or len(timestamps) <= 500:
    driver.execute_script("window.scrollBy(0,500)")
    time.sleep(2)  # Give time for comments to load

    try:
        comment_tags = driver.find_elements(By.XPATH, '//yt-formatted-string[@id="content-text"]')
        for element in comment_tags:
            comments.append(element.text)
    except NoSuchElementException as e:
        comments.append("-")
        print("NoSuchElementException", e)

    try:
        upvote_tags = driver.find_elements(By.XPATH, '//span[@id="vote-count-middle"]')
        for vote in upvote_tags:
            upvotes.append(vote.text)
    except NoSuchElementException as e:
        upvotes.append("-")
        print("NoSuchElementException", e)

    try:
        time_tags = driver.find_elements(By.XPATH, '//a[@class="yt-simple-endpoint style-scope yt-formatted-string"]')
        for times in time_tags:
            timestamps.append(times.text)
    except NoSuchElementException as e:
        timestamps.append("-")
        print("NoSuchElementException", e)

print("Number of comments:", len(comments))
print("Number of upvotes:", len(upvotes))
print("Number of timestamps:", len(timestamps))

# Close the driver
driver.close()



```

```python
import re

def find_oneplus(name):
    branded_name=[]
    for brand in name:
        brands=re.findall(r"^OnePlus",brand)
        branded_name.append(brands)
    return branded_name
        
    
title=find_oneplus(brand_name)
print(title)
    
```

```python

def extracts_mobile_name(smart):
    mobile_name=[]
    for mob in smart:
        mobiles=re.findall("\s(.*)5G",mob)
        mobile_name.append(mobiles)
    return mobile_name

smartphone=extracts_mobile_name(brand_name)

print(smartphone)
```

```python
def extracts_color_name(rang):
    colour_name=[]
    for col in rang:
        colours=re.findall("\((.*),",col)
        colour_name.append(colours)
    return colour_name

texture=extracts_color_name(brand_name)

print(texture)
```

```python
def extract_ram(fast):
    ram_put=[]
    for r in fast:
        rams=re.findall(r"(\b\d+)\s*GB\s*RAM",r)
        ram_put.append(rams)
    return ram_put

random=extract_ram(ram)

print(random)
```

```python
def extract_rom(read):
    rom_put=[]
    for i in read:
        roms=re.findall(r"(\b\d+)\s*GB\s*ROM",i)
        rom_put.append(roms)
    return rom_put

memory=extract_rom(rom)

print(memory)
```

Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import WebDriverException, ElementClickInterceptedException, ElementNotInteractableException, StaleElementReferenceException, NoSuchElementException, TimeoutException
import time
import requests
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

```python
driver=webdriver.Chrome()
driver.get("https://www.forbes.com/?sh=4c28931c2254")
time.sleep(2)

list_click=driver.find_element(By.XPATH,'//div[@class="_8FT-x3t4"]')
list_click.click()
time.sleep(2)

try:
    billion_in=driver.find_element(By.XPATH,'/html/body/div[1]/header/nav/div[1]/div[1]/div/div[2]/ul/li[2]/div[1]/span')
    billion_in.click()
except NoSuchElementException as e:
    print("NoSuchElementException",e)
time.sleep(2)

try:
    world=driver.find_element(By.XPATH,'//li[@class="TjJgrPSg _2bNo56RE secondary"]')
    world.click()
except ElementClickInterceptedException as e:
    print("ElementClickInterceptedException",e)
time.sleep(2) 
```

```python
rank=[]
name=[]
net_worth=[]
age=[]
citizenship=[]
source=[]
industry=[]

try:
    rank_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][1]')
    for i in rank_tag:
        rank.append(i.text)
except Exception as e:
    print("Exception",e)
try:
    name_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][2]')
    for i in name_tag:
        name.append(i.text)
except Exception as e:
    print("Exception",e)
try:
    net_worth_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][3]')
    for i in net_worth_tag:
        net_worth.append(i.text)
except Exception as e:
    print("Exception",e)
try:
    age_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][4]')
    for i in age_tag:
        age.append(i.text)
except Exception as e:
    print("Exception",e)
try:
    citizenship_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][5]')
    for i in citizenship_tag:
        citizenship.append(i.text)
except Exception as e:
    print("Exception",e)
    
time.sleep(2)

try:
    source_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][6]')
    for i in source_tag:
        source.append(i.text)
except Exception as e:
    print("Exception",e)
    
try:
    industry_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][7]')
    for i in industry_tag:
        industry.append(i.text)
except Exception as e:
    print("Exception",e)        

    
driver.close()    
```

```python
df_rich=pd.DataFrame({"rank":rank,"name":name,"net_worth":net_worth,"age":age,"citizenship":citizenship,"source":source,"industry":industry})

df_rich
```
