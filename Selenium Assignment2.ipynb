Q2: Write a python program to scrape data for ““Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data. This task will be done in following steps:
 1. First get the webpage https://www.naukri.com/ 
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field. 
3. Then click the searchbutton.
 4. Then scrape the data for the first 10 jobs results you get. 
5. Finally create a dataframe of the scraped data. Note: All of the above steps have to be done in code. No step is to be done manually. 


```python
import pandas as pd 
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.naukri.com/")
```

```python
designation1=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[1]/div/div/div/div[1]/div/input')
designation1.send_keys('Data Analyst')
```

```python
location1=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location1.send_keys('Bangalore')
```

```python
search1=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search1.click()
```

```python
role_title=[]
role_location=[]
org_name=[]
req_exp=[]
```

```python
role_tags=driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in role_tags:
    role=i.text
    role_title.append(role)
    

region_tags=driver.find_elements(By.CLASS_NAME,"locWdth")
for i in region_tags:
    region=i.text
    role_location.append(region)
    
import re
area=[]
for i in role_location:
    find=re.findall(r'\bBangalore|Bengaluru\b',i)
    area.append(find)
    
        
org_tags=driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[1]')
for i in org_tags:
    org=i.text
    org_name.append(org)
    
req_exp_tags=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in req_exp_tags:
    exp=i.text
    req_exp.append(exp)
    
req_exp
```

```python
print(len(role_title),len(area),len(org_name),len(req_exp))
```

```python
role_location
```

```python
df1=pd.DataFrame({"job_title":role_title,"job_location":area,"company_name":org_name,"req_exp":req_exp})
df1.head(10)
```

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
driver2=webdriver.Chrome()
```

```python
driver2.get("https://www.flipkart.com/")
```

```python
product=driver2.find_element(By.CLASS_NAME,"_3704LK")
product.send_keys("sunglasses")
```

```python
search2=driver2.find_element(By.CLASS_NAME,"L0Z3Pu")
search2.click()
```

```python
brand=[]
product_desc=[]
price=[]
```

```python
brand_tags=driver2.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
for i in brand_tags:
    item=i.text
    brand.append(item)
    
product_tags=driver2.find_elements(By.XPATH,'//div[@class="_2B099V"]/a[1]')
for i in product_tags:
    desc=i.text
    product_desc.append(desc)
    
price_tags=driver2.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
for i in price_tags:
    cost=i.text
    price.append(cost)
    
```

```python
print(len(brand),len(product_desc),len(price))
```

```python
product_title=[]

start=0
end=3


for page in range(start,end):
    titles=driver2.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in titles:
        product_title.append(i.text)
    next_button=driver2.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    next_button.click()
    time.sleep(3)
    
    
```
